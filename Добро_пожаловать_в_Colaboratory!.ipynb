```
import numpy as np
import matplotlib.pyplot as plt

# Загружаем данные о населении и случаях насилия в областях Кыргызстана
# Формат: [Год, Область, Число случаев насилия, Население]
data = np.array([
    [2018, 'Бишкек', 1200, 1100000],
    [2018, 'Ошская', 800, 1300000],
    [2018, 'Чуйская', 600, 936000],
    [2018, 'Таласская', 400, 246000],
    [2018, 'Иссык-Кульская', 500, 438000],
    [2018, 'Нарынская', 300, 290000],
    [2018, 'Баткенская', 700, 548000],
    [2018, 'Джалал-Абадская', 1000, 1238000],
    [2019, 'Бишкек', 1300, 1100000],
    [2019, 'Ошская', 850, 1300000],
    [2019, 'Чуйская', 650, 936000],
    [2019, 'Таласская', 420, 246000],
    [2019, 'Иссык-Кульская', 520, 438000],
    [2019, 'Нарынская', 350, 290000],
    [2019, 'Баткенская', 750, 548000],
    [2019, 'Джалал-Абадская', 1100, 1238000],
    [2020, 'Бишкек', 1500, 1100000],
    [2020, 'Ошская', 950, 1300000],
    [2020, 'Чуйская', 700, 936000],
    [2020, 'Таласская', 450, 246000],
    [2020, 'Иссык-Кульская', 600, 438000],
    [2020, 'Нарынская', 400, 290000],
    [2020, 'Баткенская', 800, 548000],
    [2020, 'Джалал-Абадская', 1200, 1238000],
    [2021, 'Бишкек', 1700, 1100000],
    [2021, 'Ошская', 1100, 1300000],
    [2021, 'Чуйская', 800, 936000],
    [2021, 'Таласская', 480, 246000],
    [2021, 'Иссык-Кульская', 650, 438000],
    [2021, 'Нарынская', 450, 290000],
    [2021, 'Баткенская', 850, 548000],
    [2021, 'Джалал-Абадская', 1300, 1238000],
    [2022, 'Бишкек', 1900, 1100000],
    [2022, 'Ошская', 1300, 1300000],
    [2022, 'Чуйская', 900, 936000],
    [2022, 'Таласская', 500, 246000],
    [2022, 'Иссык-Кульская', 700, 438000],
    [2022, 'Нарынская', 500, 290000],
    [2022, 'Баткенская', 900, 548000],
    [2022, 'Джалал-Абадская', 1400, 1238000]
])

# Извлечение уникальных годов и областей
years = np.unique(data[:, 0].astype(int))
regions = np.unique(data[:, 1])

# Анализ: вычисление общего числа случаев по годам
cases_by_year = []
for year in years:
    total_cases = data[data[:, 0] == str(year), 2].astype(int).sum()
    cases_by_year.append(total_cases)

# Анализ: вычисление среднего числа случаев насилия на 100,000 населения по областям
cases_per_100k_by_region = {}
for region in regions:
    region_data = data[data[:, 1] == region]
    cases_per_100k = (region_data[:, 2].astype(int) / region_data[:, 3].astype(int) * 100000).mean()
    cases_per_100k_by_region[region] = cases_per_100k

# Вывод статистики
print("Анализ данных:")
print(f"Общее число случаев по годам: {dict(zip(years, cases_by_year))}")
print("Среднее число случаев на 100,000 населения по областям:")
for region, cases in cases_per_100k_by_region.items():
    print(f"{region}: {cases:.2f}")

# Визуализация
# 1. График общего числа случаев по годам
plt.figure(figsize=(10, 5))
plt.plot(years, cases_by_year, marker='o', label='Общее число случаев')
plt.title('Динамика насилия по годам')
plt.xlabel('Год')
plt.ylabel('Число случаев')
plt.legend()
plt.grid(True)
plt.show()

# 2. Гистограмма среднего числа случаев на 100,000 населения по областям
plt.figure(figsize=(10, 5))
plt.bar(cases_per_100k_by_region.keys(), cases_per_100k_by_region.values(), color='orange')
plt.title('Среднее число случаев на 100,000 населения по областям')
plt.xlabel('Область')
plt.ylabel('Среднее число случаев (на 100,000)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
